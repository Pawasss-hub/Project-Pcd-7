```python
import cv2 as cv
import os
import shutil

def convert_dataset_to_grayscale(src_dir, dst_dir):
    if os.path.exists(dst_dir):
        shutil.rmtree(dst_dir)

    shutil.copytree(src_dir, dst_dir, ignore=shutil.ignore_patterns('*.*'))

    for root, dirs, files in os.walk(src_dir):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
                src_file_path = os.path.join(root, file)
                relative_path = os.path.relpath(root, src_dir)
                dst_file_folder = os.path.join(dst_dir, relative_path)
                dst_file_path = os.path.join(dst_file_folder, file)

                img = cv.imread(src_file_path)
                if img is None:
                    print(f"WARNING: Failed to read image: {src_file_path}")
                    continue
                
                gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
                cv.imwrite(dst_file_path, gray_img)
                print(f'Converted {src_file_path} to grayscale -> {dst_file_path}')

convert_dataset_to_grayscale('dataset', 'dataset_gray')
```
